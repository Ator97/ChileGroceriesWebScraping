```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
f= open("chileGroseries.csv","a")
f.write("Pais|Tienda|SKU|Nombre Producto|Descriocion|Atributo\n")
```

```python
#Solucion de bebidas de lider

driver = webdriver.Chrome()
driver.implicitly_wait(10)

driver.get('https://www.lider.cl/supermercado/category/Bebidas-Licores/Bebidas/Bebidas-Light-y-Zero/_/N-184btft?N=&No=0&Nrpp=80')

myDynamicElement = driver.find_elements_by_xpath('//span[@class="reference-code"]')
ewr2             = driver.find_elements_by_xpath('//span[@class="product-name"]')
ewr3             = driver.find_elements_by_xpath('//span[@class="product-description js-ellipsis"]')
ewr4             = driver.find_elements_by_xpath('//span[@class="product-attribute"]') 

for i in range(0, len(myDynamicElement)):
    f.write("Chile|Lider|"+ myDynamicElement[i].text.split()[1][:-1]+"|"+ewr2[i].text+"|"+ewr3[i].text+"|"+ewr4[i].text+"\n")
driver.close()
```

```python
#Solucion de Refrescos de Tottus

driver = webdriver.Chrome()
driver.implicitly_wait(10)

driver.get('https://www.tottus.cl/tottus/browse/Bebidas-Light-y-Zero/114.3.2')

ewr3             = driver.find_elements_by_xpath('//div[@class=" item-product-caption"]')

for i in range(0, len(ewr3)):
    f.write("Chile|Tottus|"+
    re.findall("_[\d]+",ewr3[i].get_attribute("id"))[0][1:]+"|"+
    ewr3[i].text.split('\n')[1]+"|"+
    ewr3[i].text.split('\n')[0]+"|"+
    ewr3[i].text.split('\n')[2]+"\n")
    
driver.close()
#f.write("Pais|Tienda|SKU|Nombre Producto|Descripcion|Atributo\n")
```
