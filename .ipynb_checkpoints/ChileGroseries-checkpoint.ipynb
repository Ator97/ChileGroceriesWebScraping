```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
f= open("chileGroseries06122019.txt","a")
#f.write("Pais|Tienda|SKU|Marca|Descripcion|Atributo\n")
```

```python
#Solucion de bebidas de lider

driver = webdriver.Chrome()
driver.implicitly_wait(10)

driver.get('https://www.lider.cl/supermercado/category/Bebidas-Licores/Bebidas/Bebidas-Light-y-Zero/_/N-184btft?N=&No=0&Nrpp=80')

myDynamicElement = driver.find_elements_by_xpath('//span[@class="reference-code"]')
ewr2             = driver.find_elements_by_xpath('//span[@class="product-name"]')
ewr3             = driver.find_elements_by_xpath('//span[@class="product-description js-ellipsis"]')
ewr4             = driver.find_elements_by_xpath('//span[@class="product-attribute"]') 

for i in range(0, len(myDynamicElement)):
    f.write(
    #print(
        "Chile|Lider|"+ myDynamicElement[i].text.split()[1][:-1]+"|"
        +ewr2[i].text+"|"+ewr3[i].text+"|"+ewr4[i].text+"\n")

driver.close()
```

```python
#Solucion de Galletas-y-Colaciones-Dulces de lider

driver = webdriver.Chrome()
driver.implicitly_wait(10)

for j in range(1,6):
    driver.get('https://www.lider.cl/supermercado/category/Desayunos-y-Panader%C3%ADa/Galletas-y-Colaciones-Dulces/_/N-pbmgle?No=320&isNavRequest=Yes&Nrpp=80&page='+str(j))

    myDynamicElement = driver.find_elements_by_xpath('//span[@class="reference-code"]')
    ewr2             = driver.find_elements_by_xpath('//span[@class="product-name"]')
    ewr3             = driver.find_elements_by_xpath('//span[@class="product-description js-ellipsis"]')
    ewr4             = driver.find_elements_by_xpath('//span[@class="product-attribute"]') 

    for i in range(0, len(myDynamicElement)):
        f.write(
        #print(
            "Chile|Lider|"+ myDynamicElement[i].text.split()[1][:-1]+"|"
            +ewr2[i].text+"|"+ewr3[i].text+"|"+ewr4[i].text+"\n")
        
driver.close()
```

```python
#Solucion de Refrescos de Tottus

driver = webdriver.Chrome()
driver.implicitly_wait(10)

driver.get('https://www.tottus.cl/tottus/browse/Bebidas-Light-y-Zero/114.3.2')

ewr3             = driver.find_elements_by_xpath('//div[@class=" item-product-caption"]')

for i in range(0, len(ewr3)):
    f.write(
    #print(
    "Chile|Tottus|"+
    re.findall("_[\d]+",ewr3[i].get_attribute("id"))[0][1:]+"|"+
    ewr3[i].text.split('\n')[1]+"|"+
    ewr3[i].text.split('\n')[0]+"|"+
    ewr3[i].text.split('\n')[2]+"\n")
    
driver.close()
```

```python
import re 
import json

import requests 
from bs4 import BeautifulSoup 
for i in range(1,7): 
    URL = "https://www.jumbo.cl/bebidas-aguas-y-jugos/bebidas-gaseosas?o=OrderByTopSaleDESC&page="+str(i)
    r = requests.get(URL) 
  
    soup = BeautifulSoup(r.content, 'html5lib')
    a = str(soup.script)[31:-10].replace("\\","").replace("[\"{","[{").replace("}\"]","}]")
    a = re.sub('"imageTag":"<img src[=~\/\-#.?"\w\d\s]+>",','"imageTag":"",',a)
    b = json.loads(a[1:-1])
    #print(
    #b['plp']['plp_products']['data'][0],
    for j in range(0 , len(b['plp']['plp_products']['data']) ):
        f.write(
        #print(
            'Chile|Jumbo|'+
        b['plp']['plp_products']['data'][j]['productId']+"|"+
        b['plp']['plp_products']['data'][j]['brand']+"|"+
        b['plp']['plp_products']['data'][j]['productName']+"|"+
        str(b['plp']['plp_products']['data'][j]['ProductData'][0]['unit_multiplier_un'])+" "+
        b['plp']['plp_products']['data'][j]['ProductData'][0]['measurement_unit_un']+"\n"
        )
#Chile|Tottus|20403248|MARCO POLO|Papas Fritas Caseras|250 gr
```

```python
URL = "https://www.telemercados.cl/bebidas-y-licores/bebidas/bebidas-regulares?PS=20"
r = requests.get(URL) 
soup = BeautifulSoup(r.content, 'html5lib')

a = soup.findAll('li', layout="7d4e0d7b-0d3e-4aff-9e18-445f33b759d9")
for i in range(0 , len(a)):
    if ',' in a[i].find_all('div')[3].find_all("a")[1]['alt']:
        a1 = a[i].find_all('div')[3].find_all("a")[1]['alt'].split(",",1)
    else:
        a1 = a[i].find_all('div')[3].find_all("a")[1]['alt'].split("+",1)
    f.write(
    #print(
        "Chile|Telemercado|"+
        str(a[i].find_all('li')[1]).split('"')[4]+"|"+
        a1[0]+"|"+
        a1[1]+"\n"
    
)
    
```
