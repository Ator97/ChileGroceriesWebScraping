```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import pandas as pd
```

```python
writer = pd.ExcelWriter('ChileGroseries.xlsx', engine='xlsxwriter')
```

```python
#Solucion de bebidas de lider

driver = webdriver.Chrome()
driver.implicitly_wait(10)

driver.get('https://www.lider.cl/supermercado/category/Bebidas-Licores/Bebidas/Bebidas-Light-y-Zero/_/N-184btft?N=&No=0&Nrpp=80')

myDynamicElement = driver.find_elements_by_xpath('//span[@class="reference-code"]')
ewr2             = driver.find_elements_by_xpath('//span[@class="product-name"]')
ewr3             = driver.find_elements_by_xpath('//span[@class="product-description js-ellipsis"]')
ewr4             = driver.find_elements_by_xpath('//span[@class="product-attribute"]') 
l = []
for i in range(0, len(myDynamicElement)):
    l.append(["Chile","Lider",myDynamicElement[i].text.split()[1][:-1],
    ewr2[i].text,ewr3[i].text,ewr4[i].text])
    
df1 = pd.DataFrame(l,
        columns=["Pais","Tienda","SKU","Marca","Descripcion","Atributo"])
df1.to_excel(writer, sheet_name='Lider Bebidas')
driver.close()
```

```python
import re 
import json

import requests 
from bs4 import BeautifulSoup 
l = []
for i in range(1,7): 
    URL = "https://www.jumbo.cl/bebidas-aguas-y-jugos/bebidas-gaseosas?o=OrderByTopSaleDESC&page="+str(i)
    r = requests.get(URL) 
  
    soup = BeautifulSoup(r.content, 'html5lib')
    a = str(soup.script)[31:-10].replace("\\","").replace("[\"{","[{").replace("}\"]","}]")
    a = re.sub('"imageTag":"<img src[=~\/\-#.?"\w\d\s]+>",','"imageTag":"",',a)
    b = json.loads(a[1:-1])
    for j in range(0 , len(b['plp']['plp_products']['data']) ):
        l.append(['Chile','Jumbo',b['plp']['plp_products']['data'][j]['productId'],
        b['plp']['plp_products']['data'][j]['brand'],b['plp']['plp_products']['data'][j]['productName'],
        str(b['plp']['plp_products']['data'][j]['ProductData'][0]['unit_multiplier_un'])+" "+
        b['plp']['plp_products']['data'][j]['ProductData'][0]['measurement_unit_un']])
df1 = pd.DataFrame(l,
        columns=["Pais","Tienda","SKU","Marca","Descripcion","Atributo"])
df1.to_excel(writer, sheet_name='Jumbo Bebidas')
```

```python
URL = "https://www.telemercados.cl/bebidas-y-licores/bebidas/bebidas-regulares?PS=15"
r = requests.get(URL) 
soup = BeautifulSoup(r.content, 'html5lib')
l = []
a = soup.findAll('li', layout="7d4e0d7b-0d3e-4aff-9e18-445f33b759d9")
for i in range(0 , len(a)):
    if ',' in a[i].find_all('div')[3].find_all("a")[1]['alt']:
        a1 = a[i].find_all('div')[3].find_all("a")[1]['alt'].split(",",1)
    else:
        a1 = a[i].find_all('div')[3].find_all("a")[1]['alt'].split("+",1)
    l.append(["Chile","Telemercado",str(a[i].find_all('li')[1]).split('"')[4],
    a1[0],a1[1]])
df1 = pd.DataFrame(l,
        columns=["Pais","Tienda","SKU","Marca y Descripcion","Atributo"])
df1.to_excel(writer, sheet_name='Telemercado Bebidas')
```
